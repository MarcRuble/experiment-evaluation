# Randomize
Load a dataset as a prefab, randomize selected columns and save as `csv`.

```python
import pandas as pd
import numpy as np

# Reads a dataset and saves a randomized version.
# file: string path
# columns: list of tuples (column name, range of values)
def read_and_randomize(file, columns):
    df = pd.read_csv(file, encoding="latin_1")
    
    # for each selected column
    for (col, vals) in columns:
        rand_vals = []
        
        # generate random samples
        for line in df[col]:
            rand_vals.append(vals[np.random.randint(0, len(vals))])
        
        df[col] = pd.Series(rand_vals)
        
    index = file.find('.csv')
    file = file[:index] + '_random' + file[index:]
    df.to_csv(file, index=False)
    display(df)
        
cols = [('Q1', np.arange(1, 8)), ('Q2', np.arange(1, 8)), ('Q3', np.arange(1, 8)), ('Q4', np.arange(1, 8))]
cols.append(('Condition', (1, 2, 3, 4)))
cols.append(('Task', np.arange(1, 3)))
cols.append(('Participant', np.arange(1, 15)))
read_and_randomize("tables/results.csv", cols)
```
